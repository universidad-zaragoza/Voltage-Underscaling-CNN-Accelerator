```python
import matplotlib.pyplot as plt

# Datos para ilustrar
categories = ['Supervisado', 'No supervisado', 'Reforzado']
algorithms = ['Regresión Lineal', 'K-Means', 'Q-Learning']
description = ['Modelo que predice valores continuos.',
               'Algoritmo de agrupamiento.',
               'Método para aprender a tomar decisiones óptimas.']

# Crear la figura y los ejes
fig, ax = plt.subplots(figsize=(8, 6))

# Dibujar las categorías
ax.barh(categories, [1, 1, 1], color=['blue', 'green', 'orange'])

# Agregar los nombres de los algoritmos
for i, category in enumerate(categories):
    ax.text(0.1, i, algorithms[i], va='center', fontsize=12, fontweight='bold', color='white')

# Agregar las descripciones
for i, category in enumerate(categories):
    ax.text(1.1, i, description[i], va='center', fontsize=10)

# Estilizar los ejes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

# Ajustar el diseño
plt.tight_layout()

# Guardar la imagen
plt.savefig('machine_learning_concepts.png')

# Mostrar la imagen
plt.show()
```
