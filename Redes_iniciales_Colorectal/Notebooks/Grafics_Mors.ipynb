```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Datos de ejemplo
redes = ['Red1', 'Red2', 'Red3']  # Nombres de las redes
tecnicas = ['Tecnica1', 'Tecnica2', 'Tecnica3']  # Nombres de las técnicas
voltajes = ['Voltaje1', 'Voltaje2', 'Voltaje3']  # Nombres de los voltajes

# Precisión de cada técnica para cada voltaje en cada red (datos de ejemplo)
precisiones = np.random.rand(len(redes), len(tecnicas), len(voltajes))

# Crear el gráfico apilado
fig, ax = plt.subplots()

# Configurar los patrones para las técnicas
patrones = ['/', 'o', '+']

# Configurar las barras para cada técnica en cada red
ancho_barra = 0.2
for i, tecnica in enumerate(tecnicas):
    for j, red in enumerate(redes):
        x = np.arange(len(voltajes)) + i * ancho_barra
        ax.bar(x, precisiones[j, i, :], width=ancho_barra, label=f'{tecnica} - {red}', hatch=patrones[i])

# Personalizar el gráfico
ax.set_ylabel('Precisión')
ax.set_title('Precisión de técnicas por red y voltaje')
ax.set_xticks(np.arange(len(redes)) + 0.5 * len(tecnicas) * ancho_barra)
ax.set_xticklabels(redes)

# Crear la leyenda con los patrones adecuados
legend_elements = [Patch(facecolor='white', edgecolor='black', label=f'{tec} - {red}', hatch=pat) for tec in tecnicas for red in redes for pat in patrones]
ax.legend(handles=legend_elements, loc='upper center', bbox_to_anchor=(0.5, -0.1), fancybox=True, shadow=True, ncol=len(tecnicas))

# Mostrar el gráfico
plt.show()
```
